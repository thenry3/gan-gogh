```python
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import re

%load_ext autoreload
%autoreload 2
```

```python
with open("./losses.txt") as file:
    data = file.read()
```

```python
batch_losses = []
per_losses = []
content_losses = []
style_losses = []
for line in data.split("\n"):
    if (str.isspace(line)):
        continue
    batch = re.search(r"Batch loss: \[([A-Za-z0-9_.]+)\]", line)
    batch_losses.append( float(batch.group(1)) )
    per = re.search(r"Perceptual: \[([A-Za-z0-9_.]+)\]", line)
    per_losses.append( float(per.group(1)) )
    content = re.search(r"Content: \[([A-Za-z0-9_.]+)\]", line)
    content_losses.append( float(content.group(1)) )
    style = re.search(r"Style: \[([A-Za-z0-9_.]+)\]", line)
    style_losses.append( float(style.group(1)) )
    
```

```python
plt.title("VGG19 Loss")
plt.plot(batch_losses, color = "purple", label = "Batch Loss")
plt.plot(content_losses, label = "Content Loss")
plt.plot(style_losses, label = "Style Loss")
plt.ylabel("Loss")
plt.xlabel("Step")
plt.legend()
```
