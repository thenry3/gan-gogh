```python
import numpy as np
import matplotlib.pyplot as plt
import re

%load_ext autoreload
%autoreload 2
```

```python
with open("./losses.txt") as file:
    data = file.read()
```

```python
D_items = re.findall("Loss_D:\s([-+]?\d*\.*\d+)", data)
D_losses = [float(x) for x in D_items]

G_items = re.findall("Loss_G:\s([-+]?\d*\.*\d+)", data)
G_losses = [float(x) for x in G_items]
```

```python
plt.title("DCGAN Loss")
plt.plot(D_losses, color = "purple", label = "Discriminator Loss")
plt.plot(G_losses, label = "Generator Loss")
plt.ylabel("Loss")
plt.xlabel("Step")
plt.legend()
```
